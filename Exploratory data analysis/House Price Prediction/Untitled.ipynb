```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Display all the columns of the dataframes

pd.pandas.set_option('display.max_columns', None)
```

```python
dataset = pd.read_csv('train.csv')
```

```python
dataset.head()
```

```python
# Missing vlaues

dataset.isnull()
```

```python
dataset.isnull().sum()
```

```python

features_with_na = [features for features in dataset.columns if dataset[features].isnull().sum()]
features_with_na
```

```python
for feature in features_with_na:
    print(feature, np.round(dataset[feature].isnull().mean(), 4), '% missing vlaues')
```

```python
for feature in features_with_na:
    print(feature, np.round(dataset[feature].isnull().mean() * 100, 4), '% missing values')
```

```python
for feature in features_with_na:
    if np.round(dataset[feature].isnull().mean()*100, 4) > 50:
        print(feature, np.round(dataset[feature].isnull().mean() * 100, 4), '% missing values')
```

```python
dataset.head()
```

```python
data = dataset.copy()

for feature in features_with_na:
    
    data[feature] = np.where(data[feature].isnull(), 1, 0)
    
    data.groupby(feature)['SalePrice'].median()
    print(data)
```
