```python
# data_food describing the overall food order

data_food = pd.read_csv(r"D:\iiit notes\Internship\23 season 14th jun-2021\EDA\food.csv", header = 0)
data_food.head()
```

```python
# Merge all the three files into a single data file 

food_data = pd.merge(data_food, data_center, on = 'center_id')
food_data = pd.merge(food_data, data_meal, on = 'meal_id')
food_data.head()
```

```python
# Size of the data after merging three datafiles

food_data.shape
```

```python
# Creating Total_price varible from checkout_price and num_orders 

food_data['Total_Price'] = (food_data['checkout_price']* food_data['num_orders'])
```
