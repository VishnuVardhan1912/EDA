```python
# Create cols1 for creating dummy variables

cols1 = ['Agent_Type','Gender','Policy_Type']

# Create cols2 for scaling the data - normalization

cols2 = ['Num_Drivers','Num_Vehicles','State','Age_Range', 'Driving_Exp_Range', 'Prev_Accidents',
       'Prev_Citations', 'Marital_Status', 'Education', 'Sal_Range',
       'Coverage', 'Veh_Usage', 'Annual_Miles_Range', 'Vehicle_Cost_Range']
```

```python
# Create dummy variable for all col1 columns

AutoIns = pd.get_dummies(AutoIns, columns=cols1)
AutoIns.head().T
```

```python
AutoIns.shape
```

```python
# Identify the dependent and Target variables

IndepVar=[]
for col in AutoIns.columns:
    if col != 'GW_Premium':
        IndepVar.append(col)
        
TargetVar = 'GW_Premium'

x = AutoIns[IndepVar]
y = AutoIns[TargetVar]
```

```python
# Splitting the dataset into train and test

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=6)

# Copy the test data to back-up file

x_test_F1 = x_test.copy()
```

```python
# Scale the cols2 columns by using MinMaxScaler

from sklearn.preprocessing import MinMaxScaler

mmscaler = MinMaxScaler(feature_range=(0, 1))

x_train[cols2] = mmscaler.fit_transform(x_train[cols2])
x_train = pd.DataFrame(x_train)

x_test[cols2] = mmscaler.fit_transform(x_test[cols2])
x_test = pd.DataFrame(x_test)
```

# Multi Linear Regression

```python
# Train the 'Multi Linear Regression' algorithm and build the model

from sklearn.linear_model import LinearRegression

AutoInsMLR = LinearRegression(fit_intercept=True, normalize=False, copy_X=True, n_jobs=None, positive=False)

# fit the model 

AutoInsMLR.fit(x_train, y_train)

# Predict the model with test dataset

y1_pred = AutoInsMLR.predict(x_test)

# Evalution metrics for Regression analysis

from sklearn import metrics

print('Mean Absolute Error (MAE):', round(metrics.mean_absolute_error(y_test, y1_pred),3))  
print('Mean Squared Error (MSE):', round(metrics.mean_squared_error(y_test, y1_pred),3))  
print('Root Mean Squared Error (RMSE):', round(np.sqrt(metrics.mean_squared_error(y_test, y1_pred)),3))
print('Mean Absolute Percentage Error (MAPE):', round(metrics.mean_absolute_percentage_error(y_test, y1_pred) * 100, 3), '%')
print('R2_score:', round(metrics.r2_score(y_test, y1_pred),4))

# Calculate Adjusted R squared value

r_squared = round(metrics.r2_score(y_test, y1_pred),4)
adjusted_r_squared = round(1 - (1-r_squared)*(len(y)-1)/(len(y)-x.shape[1]-1),4)
print('Adjusted R2_Score', adjusted_r_squared)
```
