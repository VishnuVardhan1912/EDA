```python
# Load the Auto_Insurance dataset

AutoIns = pd.read_excel(r"D:\iiit notes\Programming\AI\Internship practice\Capstone project - 02\Auto_Insurance_Data V 1.0.xlsx",sheet_name=0, header=0)

# Display first five records 

AutoIns.head()
```

```python
# Adding new columns as year, month, week Day and week Day for Bind_Dt

AutoIns['B_year'] = AutoIns['Bind_Dt'].dt.year
AutoIns['B_Month'] = AutoIns['Bind_Dt'].dt.month
AutoIns['B_WeekNum'] = AutoIns['Bind_Dt'].dt.weekofyear
AutoIns['B_Day'] = AutoIns['Bind_Dt'].dt.weekday
```

```python
# Drop the variables which are not consider for time series analysis

AutoIns = AutoIns.drop(['Rec_ID', 'Agent_Type','Days_Process_Bind','State','Policy_Type', 'NW_Premium', 'Taxes','Num_Drivers',
                        'Num_Vehicles', 'Gender', 'Age', 'Driving_Exp', 'Prev_Accidents','Prev_Citations', 'Marital_Status', 
                        'Education', 'Sal_Range','Coverage', 'Veh_Usage', 'Annual_Miles_Range', 'Vehicle_Cost_Range', 
                        'Claims_Amount_Range', 'Unnamed: 28'], axis=1)
```

```python
# Display the dataset information after deletion of variables (not impacting our solution) 

AutoIns.info()
```

```python
# To understand the New Policy bind trend with 6 months predictions

AutoIns_File1 = psql.sqldf("select B_Month, B_year, count(B_Month) as Total_Count, \
                            sum(GW_Premium) as Total_Policy_Premium \
                            from AutoIns \
                            group by B_Month, B_year \
                            order by B_Month, B_year")
AutoIns_File1.head()
```
