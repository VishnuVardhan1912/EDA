```python
# importing the libraries

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
plt.style.use('seaborn')

import pandasql as psql 
```

```python
# data_meal describing meals
data_meal = pd.read_csv(r"D:\iiit notes\Internship\23 season 14th jun-2021\EDA\meal_info.csv", header=0)
data_meal.head()
```

```python
# data_center describing region

data_center = pd.read_csv(r"D:\iiit notes\Internship\23 season 14th jun-2021\EDA\fulfilment_center_info.csv", header =0)
data_center.head()
```

```python
# data_food describing food

data_food = pd.read_csv(r"D:\iiit notes\Internship\23 season 14th jun-2021\EDA\food.csv", header=0)
data_food.head()
```

```python
# Merge all the three data files into a single data file

food_master = pd.merge(data_food, data_center, on='center_id')
food_master = pd.merge(food_master, data_meal, on='meal_id')
food_master.head()
```

```python
# Create Total_price (Total_price = checkout_price * num_orders)

food_master['Total_Price'] = food_master['checkout_price'] * food_master['num_orders']
food_master.head()
```

```python
food_master['op_area'] = (food_master['op_area'] * 1000)
food_master['op_area'] = food_master['op_area'].astype(int)
food_master.head()
```

```python
# Calculate the Average Revenue per SQFT

food_master['Ave_Revenue_SQFT'] = (food_master['Total_Price'] / food_master['op_area'])
food_master.head()
```

```python
food_master.to_csv("D:/iiit notes/Internship/25 season 16th jun-2021/food_master.csv")
```
